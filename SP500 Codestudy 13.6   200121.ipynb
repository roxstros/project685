```python
start_time = datetime.now() 
time.sleep(3)   


df[["TR", "N", "vlag_sep", "NPP", "nEMA", "FL", "SL", "Sep", "TrendDay", "FE_FL", "AE_FL", "SP", 
    "Pos_St", "Pos_End", "a", "b", "c", "En_price1", "En_price2", "En_price3", "En_price4", "Ex_price1", 
    "Ex_price2", "Ex_price3", "Ex_price4", "P_L", "Capital"]] = df[["TR", "N", "vlag_sep", "NPP", "nEMA", 
                        "FL", "SL", "Sep", "TrendDay", "FE_FL", "AE_FL", "SP", "Pos_St", "Pos_End", "a", 
                        "b", "c", "En_price1", "En_price2", "En_price3", "En_price4", "Ex_price1", 
                        "Ex_price2", "Ex_price3", "Ex_price4", "P_L", "Capital"]].apply(pd.to_numeric).round(sigfigs)

tick = 0.25
tick_val = 12.50
# Capital = float(25000)
# df["Capital"] = pd.to_numeric(df["Capital"], errors = "coerce")

   
for rownum in range(x+1, 2500):        
    if df.TrendDay.loc[rownum] == 0 and df.TrendDir.loc[rownum] == "Up":          
        Pos_St = ""
        df.Pos_St.loc[rownum] = Pos_St
        SP = df.high.loc[rownum]                            # assign the 'high' slice to the SP variable
        df.SP.loc[rownum] = SP                              # equate the SP slice with the SP variable
        df.Capital.loc[rownum] = df.Capital.loc[rownum-1]
        rownum += 1

        while df.TrendDay.loc[rownum] > 0:     # this loop finds initial entry, records initial P_L, then advances to next day
            if df.high.loc[rownum] > SP:
                En_price1 = SP + tick
                Counter1 = 1
                Pos_St = 1                                
                Pos_End = 1
                df.En_price1.loc[rownum] = En_price1
                df.Counter1.loc[rownum] = Counter1   
                df.Pos_St.loc[rownum] = Pos_St
                df.Pos_End.loc[rownum] = Pos_End
                df.P_L.loc[rownum] = (df.close.loc[rownum] - En_price1) / tick * tick_val
                df.Capital.loc[rownum] = df.Capital.loc[rownum-1] + df.P_L.loc[rownum]
                rownum += 1
                break

            else:
                SP = df.high.loc[rownum]
                df.SP.loc[rownum] = SP
                rownum += 1  


    elif df.TrendDay.loc[rownum] == 0 and df.TrendDir.loc[rownum] == "Down":
        Pos_St = ""
        SP = df.low.loc[rownum]
        df.Pos_St.loc[rownum] = Pos_St
        df.SP.loc[rownum] = SP
        df.Capital.loc[rownum] = df.Capital.loc[rownum-1]
        rownum += 1        

        while df.TrendDay.loc[rownum] > 0:
            if df.low.loc[rownum] < SP:
                En_price1 = SP - tick
                Counter1 = 1                
                Pos_St = 1
                Pos_End = 1
                df.En_price1.loc[rownum] = En_price1
                df.Counter1.loc[rownum] = Counter1 
                df.Pos_St.loc[rownum] = Pos_St                                
                df.Pos_End.loc[rownum] = Pos_End
                df.P_L.loc[rownum] = (En_price1 - df.close.loc[rownum]) / tick * tick_val  
                df.Capital.loc[rownum] = df.Capital.loc[rownum-1] + df.P_L.loc[rownum]                
                rownum += 1
                break
            else:
                SP = df.low.loc[rownum]
                df.SP.loc[rownum] = SP
                rownum += 1


print('Time elapsed (hh:mm:ss.ms) {}'.format(datetime.now() - start_time))  
```

```python
"""  UPtrend Functions  """
"""
def uptrend_zeros():
    if df.NPP.loc[rownum] <= 3.00:
        if df.TrendDir.loc[rownum] == "Up":                                  # Position = 0, Uptrend
            if df.c_type.loc[rownum] == "o" or df.c_type.loc[rownum] == "d-o":      #  low precedes High
                if df.high.loc[rownum] > df.c.loc[rownum]: 
                    if df.low.loc[rownum] < df.b.loc[rownum]:          # u0f: Re-enter on 2 positions then SL on 2 position           
                        u_0f()
                    elif df.low.loc[rownum] <= df.d.loc[rownum]:       # u0g: Re-enter on 2 positions then TP on 2 then set RecanTP or RecanSL          
                        u_0g()
                    elif df.low.loc[rownum] > df.d.loc[rownum]:        # u0h: No Re-entry, stay RecanTP or RecanSL =True          
                        u_0h()

                elif df.high.loc[rownum] < df.c.loc[rownum]: 
                    if df.low.loc[rownum] < df.b.loc[rownum]:          # u0j: Re-enter on 2 positions then SL on 2 positions           
                        u_0j
                    elif df.low.loc[rownum] <= df.d.loc[rownum]:       # u0k: Re-enter on 2 positions then CARRY 2 positions          
                        u_0k()
                    elif df.low.loc[rownum] > df.d.loc[rownum]:        # u0m: No Re-entry, stay RecanTP or RecanSL =True                             
                        u_0m()

            elif df.c_type.loc[rownum] == "d" or df.c_type.loc[rownum] == "d-d":   #  High precedes Low
                if df.high.loc[rownum] > df.c.loc[rownum]: 
                    if df.low.loc[rownum] < df.b.loc[rownum]:           # u0n: Re-enter on 2 positions then SL on 2 positions           
                        u_0n()
                    elif df.low.loc[rownum] <= df.d.loc[rownum]:       # u0o: Re-enter on 2 positions then CARRY 2 positions        
                        u_0o()
                    elif df.low.loc[rownum] > df.d.loc[rownum]:        # u0q: No Re-entry, stay RecanTP or RecanSL =True           
                        u_0q()

                elif df.high.loc[rownum] < df.c.loc[rownum]: 
                    if df.low.loc[rownum] < df.b.loc[rownum]:          # u0r: Re-enter on 2 positions then SL on 2 positions          
                        u_0r()
                    elif df.low.loc[rownum] <= df.d.loc[rownum]:       # u0s: Re-enter on 2 positions then CARRY 2 positions 
                        u_0s()
                    elif df.low.loc[rownum] > df.d.loc[rownum]:        # u0t: No Re-entry, stay RecanTP or RecanSL =True                               
                        u_0t()

    elif df.NPP.loc[rownum] > 3.00:
        if df.TrendDir.loc[rownum] == "Up":                                  # Position = 0, Uptrend
            if df.c_type.loc[rownum] == "o" or df.c_type.loc[rownum] == "d-o":      #  low precedes High
                if df.high.loc[rownum] > df.c.loc[rownum]: 
                    
        
"""        
        
        
        


""" Uptrend, Pos_St=1      c_type: open & d-o       case: high>c, low<a&b  """
def uO_1_AP1_SL2_0():                                                             # uO_1_AP1_SL2_0:  AP then SL on 2 positions 
    global rownum
    Pos_St = df.Pos_End.loc[rownum-1]
    df.Pos_St.loc[rownum] = Pos_St
    
    """ AP """
    En_price2 = df.a.loc[rownum-1] - tick           
    df.En_price2.loc[rownum] = En_price2                                  

    """ SL on 2  """
    Ex_price3 = df.b.loc[rownum-1] - tick
    Ex_price4 = df.b.loc[rownum-1] - tick
    df.Ex_price3.loc[rownum] = Ex_price3
    df.Ex_price4.loc[rownum] = Ex_price4
    df.Counter1.loc[rownum] = ""
    df.Counter2.loc[rownum] = ""
    RecanSL = True
    Pos_End = 0
    df.Pos_End.loc[rownum] = Pos_End 
    df.RecanSL.loc[rownum] = RecanSL
    if df.c_type.loc[rownum] == "o" or df.c_type.loc[rownum] =="d-o":
        df.Trade_Tag.loc[rownum] = "uO_1_AP1_SL2_0" 
    elif df.c_type.loc[rownum] == "d" or df.c_type.loc[rownum] =="d-d":
        df.Trade_Tag.loc[rownum] = "uD_1_AP1_SL2_0"    
    P_L = ((Ex_price3 - En_price2) + (Ex_price4 - df.close.loc[rownum-1])) / tick * tick_val
    df.P_L.loc[rownum] = P_L
    df.Capital.loc[rownum] = df.Capital.loc[rownum-1] + P_L
    rownum += 1
        
    
""" Uptrend, Pos_St=1      c_type: open & d-o       case: evaluate low<a then high>c   """    
def uO_1_AP1_TP2_0():                               # u1g:  AP then TP on 2 positions then set RecanTP  
    global rownum
    Pos_St = df.Pos_End.loc[rownum-1]
    En_price2 = df.a.loc[rownum-1] - tick 
    Ex_price1 = df.c.loc[rownum-1] + tick
    Ex_price2 = df.c.loc[rownum-1] + tick
    Pos_End = 0
    RecanTP = True    
    df.Ex_price1.loc[rownum] = Ex_price1
    df.Ex_price2.loc[rownum] = Ex_price2    
    df.Counter1.loc[rownum] = ""
    df.Counter2.loc[rownum] = ""    
    df.En_price2.loc[rownum] = En_price2
    df.Pos_St.loc[rownum] = Pos_St
    df.Pos_End.loc[rownum] = Pos_End
    df.RecanTP.loc[rownum+1] = RecanTP      # next day RecanTP status is True
    df.Trade_Tag.loc[rownum] = "uO_1_AP1_TP2_0"      
    P_L = ((Ex_price1 - df.close.loc[rownum-1]) + (Ex_price2 - En_price2)) / tick * tick_val  
    df.P_L.loc[rownum] = P_L
    df.Capital.loc[rownum] = df.Capital.loc[rownum-1] + P_L
    rownum += 1
  

 """ Uptrend, Pos_St=1      c_type: open & d-o       case: evaluate low>a  then high>c   """    
def uO_1_TP1_0():                                       # case u1h:  TP on initial then set RECAN    uO_TP1_0
    global rownum
    Pos_St = 1
    Ex_price1 = df.c.loc[rownum-1] + tick
    df.Ex_price1.loc[rownum] = Ex_price1
    df.Counter1.loc[rownum] = ""
    Pos_End = 0
    RecanTP = True
    df.Pos_St.loc[rownum] = Pos_St
    df.Pos_End.loc[rownum] = Pos_End
    df.RecanTP.loc[rownum] = RecanTP  # toggle RecanTP to True
    df.Trade_Tag.loc[rownum] = "uO_1_TP1_0"    
    P_L = (Ex_price1 - df.close.loc[rownum-1]) / tick * tick_val
    df.P_L.loc[rownum] = P_L  
    df.Capital.loc[rownum] = df.Capital.loc[rownum-1] + P_L
    rownum += 1
    

""" Uptrend, Pos_St=1      c_type: open & d-o       case: high<c, low<a&b   """       
def u_1j():                                                             # u1j:  AP then SL on 2 positions 
    global rownum
    Pos_St = 1
    df.Pos_St.loc[rownum] = Pos_St
    En_price2 = df.a.loc[rownum] - tick
    df.En_price2.loc[rownum] = En_price2   
    u_1f()
    df.Trade_Tag.loc[rownum] = "u1j"     
    rownum += 1
    """ Check for Re-entry after Uptrend stop out """
    while df.TrendDay.loc[rownum] != 0:
        if df.FL.loc[rownum] > df.FL.loc[rownum-1]:
            if df.FL.loc[rownum-1] > df.FL.loc[rownum-2]:
                RE_CAN = True
                df.RE_CAN.loc[rownum] = RE_CAN
                df.Pos_St.loc[rownum] = 0
                df.RE_CAN.loc[rownum+1] = RE_CAN
                df.Pos_St.loc[rownum+1] = 0
                rownum += 1
                uptrend_zeros()      # call Pos_St = 0, RECAN=True scenario functions

            else:
                rownum += 1    
        else:
            rownum += 1
    
    
""" Uptrend, Pos_St=1      c_type: open & d-o       case: high<c, low<a   """      
def uO_1_AP1_2():                                                             # u1k:  AP then carry 2 positions
    global rownum
    Pos_St = 1
    Counter1 += 1
    En_price2 = df.a.loc[rownum] - tick
    Counter2 = 1
    df.En_price2.loc[rownum] = En_price2
    df.Counter1.loc[rownum] = Counter1
    df.Counter2.loc[rownum] = Counter2
    Pos_End = 2
    df.Pos_St.loc[rownum] = Pos_St    
    df.Pos_End.loc[rownum] = Pos_End
    if df.c_type.loc[rownum] == "o" or df.c_type.loc[rownum] =="d-o":
        df.Trade_Tag.loc[rownum] = "uO_1_AP1_2" 
    elif df.c_type.loc[rownum] == "d" or df.c_type.loc[rownum] =="d-d":
        df.Trade_Tag.loc[rownum] = "uD_1_AP1_2"       
    PL1 = (df.close.loc[rownum] - df.close.loc[rownum-1]) / tick * tick_val
    PL2 = (df.close.loc[rownum] - En_price2) / tick * tick_val
    P_L = PL1 + PL2
    df.P_L.loc[rownum] = P_L
    df.Capital.loc[rownum] = df.Capital.loc[rownum-1] + P_L
    rownum += 1

    
""" Uptrend, Pos_St=1      c_type: open & d-o       case: high<c, low>a     """  
def uO_1_1():                                                    # u1m:  No volatility impact - carry initial position
    global rownum
    Pos_St = 1
    Pos_End = 1
    Counter1 += 1
    df.Pos_St.loc[rownum] = Pos_St  
    df.Counter1.loc[rownum] = Counter1
    df.Pos_End.loc[rownum] = Pos_End 
    if df.c_type.loc[rownum] == "o" or df.c_type.loc[rownum] =="d-o":
        df.Trade_Tag.loc[rownum] = "uO_1_1" 
    elif df.c_type.loc[rownum] == "d" or df.c_type.loc[rownum] =="d-d":
        df.Trade_Tag.loc[rownum] = "uD_1_1"          
    P_L = (df.close.loc[rownum] - df.close.loc[rownum-1]) / tick * tick_val       
    df.P_L.loc[rownum] = P_L
    rownum += 1    
           
        
""" Uptrend, Pos_St=1      c_type: DARK & d-D       case: high>c, low<a&b    """           
def uD_1_TP1_0():                                       # u1n:  TP on initial position then Re-enter 2 then SL on 2
    global rownum
    Pos_St = 1
    df.Pos_St.loc[rownum] = Pos_St                    
    """ TP on initial and set RECAN=True  """
    Ex_price1 = df.c.loc[rownum-1] + tick
    df.Ex_price1.loc[rownum] = Ex_price1
    Pos_End = 0
    
    """ Check for same day Re-entry due to Low exceeding (Ex_price1 - 0.5*N)  """
    if df.low.loc[rownum] < (Ex_price1 - 0.5 * df.N.loc[rownum-1]):
        En_price1 = (Ex_price1 - 0.5 * df.N.loc[rownum-1]) - tick    # Re-enter 1st position
        En_price2 = (Ex_price1 - 0.5 * df.N.loc[rownum-1]) - tick    # Re-enter 2nd position
        RecanTP = ""      # toggle RecanTP back to off
        Pos_End = 2
        Counter1 = 1
        Counter2 = 1
        df.En_price1.loc[rownum] = En_price1
        df.En_price2.loc[rownum] = En_price2
        df.Pos_St.loc[rownum] = Pos_St
        df.Pos_End.loc[rownum] = Pos_End
        df.Counter1.loc[rownum] = Counter1
        df.Counter2.loc[rownum] = Counter2
        df.Trade_Tag.loc[rownum] = "uD_1_TP1_RE2_2"
        PL1 = Ex_price1 - df.close.loc[rownum-1]
        PL2 = (df.close.loc[rownum] - En_price1) * 2
        P_L = (PL1 + PL2) / tick * tick_val
        df.Capital.loc[rownum] = df.Capital.loc[rownum-1] + P_L
        rownum +=1
        
        """ if Low exceeds B then stop out of re-entry positions """
        if df.low.loc[rownum] < df.b.loc[rownum-1]:
            Pos_End = 0
            RecanSL = True
            Ex_price3 = df.b.loc[rownum] - tick
            Ex_price4 = df.b.loc[rownum] - tick
            Trade_Tag = "uD_1_TP1_RE2_SL2_0"
            Counter1 = ""
            Counter2 = ""
            df.Pos_End.loc[rownum] = Pos_End
            df.Ex_price3.loc[rownum] = Ex_price3
            df.Ex_price4.loc[rownum] = Ex_price4
            df.Trade_Tag.loc[rownum] = Trade_Tag
            df.Counter1.loc[rownum] = Counter1
            df.Counter2.loc[rownum] = Counter2 
            df.RecanSL.loc[rownum] = RecanSL
            P_L = (En_price1 - Ex_price3) * 2 / tick * tick_val
            df.Capital.loc[rownum] = df.Capital.loc[rownum-1] + P_L
            rownum += 1
    
    else:
        Pos_End = 0
        RecanTP = True
        Counter1 = ""
        df.RecanTP.loc[rownum] = RecanTP
        df.Ex_price1.loc[rownum] = Ex_price1     # must associate Ex_price1 with RecanTP status - always roll forward
        df.Pos_End.loc[rownum] = Pos_End
        df.Trade_Tag.loc[rownum] = "uD_1_TP1_0"
        P_L = Ex_price1 - df.close.loc[rownum-1]
        df.Capital.loc[rownum] = df.Capital.loc[rownum-1] + P_L
        rownum += 1
    
           
""" Uptrend, Pos_St=1      c_type: DARK & d-D       case: high>c, low<a      """             
def u_1p():                                                         # u1p:  TP on initial then Re-enter 2 then Carry 2
    global rownum
    Pos_St = 1
    df.Pos_St.loc[rownum] = Pos_St                    
    """ TP on initial and set RECAN=True  """
    Ex_price1 = df.c.loc[rownum] + tick
    Pos_End = 0
    RE_CAN = True
    """ Same day Re-entry due to Low exceeding 'd'  """
    En_price3 = df.d.loc[rownum] - tick   # RECAN
    En_price4 = df.d.loc[rownum] - tick   # RECAN
    Pos_End = 2                            # update Pos_End local variable 
    """ Carry 2 due to Low > 'b'   """  
    df.Pos_End.loc[rownum] = Pos_End      # final disposition of Pos_End local variable
    df.Pos_St.loc[rownum+1] = 2           # next day Pos_St = 2  
    RE_CAN = False
    df.RE_CAN.loc[rownum+1] = RE_CAN  # next day RECAN status is False
    df.Counter1.loc[rownum] = df.Counter1.loc[rownum-1] + 1
    df.Counter2.loc[rownum] = df.Counter2.loc[rownum-1] + 1
    df.Trade_Tag.loc[rownum] = "u1p"     
    P_L = ((Ex_price1 - df.close.loc[rownum]) + (df.close.loc[rownum] - En_price3) + (df.close.loc[rownum] - En_price4)) / tick * tick_val
    df.P_L.loc[rownum] = P_L     
    df.Trade_Tag.loc[rownum] = "u1p"  
    rownum += 1
    
                
""" Uptrend, Pos_St=1      c_type: DARK & d-D       case: high>c, low>a      """             
def u_1q():                                                               # u1q:  TP on initial position then set RECAN
    u_1n()                  
    df.Trade_Tag.loc[rownum] = "u1q" 

    
""" Uptrend, Pos_St=1      c_type: DARK & d-D       case: high<c, low<a&b       """                
def u_1r():                                                                # u1r:  AP then SL on 2 positions  
    u_1f()           
    df.Trade_Tag.loc[rownum] = "u1r"
    


    
""" Uptrend, Pos_St=1      c_type: DARK & d-D       case: high<c, low<a        """                
def u_1s():                                                              # u1s:  AP then carry 2 positions
    u_1k()           
    df.Trade_Tag.loc[rownum] = "u1s"        

           
""" Uptrend, Pos_St=1      c_type: DARK & d-D       case: high<c, low>a        """                
def u_1t():                                                           # u1t:  No volatility impact - carry initial position
    u_1m()                
    df.Trade_Tag.loc[rownum] = "u1t"        
     

    
    
""" Uptrend, Pos_St=2     c_type: open & d-o       case: high>c  low <a&b     """           
def u_2f():                                                                         # u2f:  SL on 2 positions
    global rownum
    Pos_St = 2
    df.Pos_St.loc[rownum] = Pos_St   
    Ex_price3 = df.b.loc[rownum] - tick
    Ex_price4 = df.b.loc[rownum] - tick
    df.Ex_price3.loc[rownum] = Ex_price3
    df.Ex_price4.loc[rownum] = Ex_price4
    df.Counter1.loc[rownum] = ""  
    df.Counter2.loc[rownum] = ""
    Pos_End = 0
    df.Pos_End.loc[rownum] = Pos_End  
    df.Pos_St.loc[rownum+1] = 0     # next day Pos_St = 0 
    df.Trade_Tag.loc[rownum] = "u2f"     
    P_L = ((Ex_price3 - df.close.loc[rownum-1]) + (Ex_price4 - df.close.loc[rownum-1])) / tick * tick_val       
    df.P_L.loc[rownum] = P_L   
    rownum += 1    
    """ Check for Re-entry after Uptrend stop out """
    while df.TrendDay.loc[rownum] != 0:
        if df.FL.loc[rownum] > df.FL.loc[rownum-1]:
            if df.FL.loc[rownum-1] > df.FL.loc[rownum-2]:
                RE_CAN = True
                df.RE_CAN.loc[rownum] = RE_CAN
                df.Pos_St.loc[rownum] = 0
                df.RE_CAN.loc[rownum+1] = RE_CAN
                df.Pos_St.loc[rownum+1] = 0
                rownum += 1
                uptrend_zeros()      # call Pos_St = 0, RECAN=True scenario functions

            else:
                rownum += 1    
        else:
            rownum += 1
    
    
""" Uptrend, Pos_St=2     c_type: open & d-o       case: high>c  low <a&b     """           
def u_2g():                                                                 # u2g:  TP on 2 positions then set RECAN
    global rownum
    Pos_St = 2
    df.Pos_St.loc[rownum] = Pos_St
    """ TP on 2 """
    Ex_price1 = df.c.loc[rownum] + tick
    Ex_price2 = df.c.loc[rownum] + tick 
    df.Ex_price1.loc[rownum] = Ex_price1
    df.Ex_price2.loc[rownum] = Ex_price2
    df.Counter1.loc[rownum] = ""  
    df.Counter2.loc[rownum] = ""
    Pos_End = 0     
    df.Pos_End.loc[rownum] = Pos_End
    df.Pos_St.loc[rownum+1] = 0     # next day Pos_St = 0  
    RE_CAN = True
    df.RE_CAN.loc[rownum+1] = RE_CAN  # next day RECAN status is True
    df.Trade_Tag.loc[rownum] = "u2g"
    P_L = ((Ex_price1 - df.close.loc[rownum-1]) + (Ex_price2 - df.close.loc[rownum-1])) / tick * tick_val       
    df.P_L.loc[rownum] = P_L
    rownum += 1                    
    
           
""" Uptrend, Pos_St=2     c_type: open & d-o       case: high>c  low >a      """           
def u_2h():                                                                      # u2h:  TP on 2 positions then set RECAN
    u_2g()       
    df.Trade_Tag.loc[rownum] = "u2h"       
           

""" Uptrend, Pos_St=2     c_type: open & d-o       case: high<c    low < a&b      """           
def u_2j():                                                                       # u2j:  SL on 2 positions
    global rownum
    Pos_St = 2
    df.Pos_St.loc[rownum] = Pos_St  
    Ex_price3 = df.b.loc[rownum] - tick
    Ex_price4 = df.b.loc[rownum] - tick
    df.Ex_price3.loc[rownum] = Ex_price3
    df.Ex_price4.loc[rownum] = Ex_price4    
    df.Counter1.loc[rownum] = ""  
    df.Counter2.loc[rownum] = ""
    Pos_End = 0
    df.Pos_End.loc[rownum] = Pos_End
    df.Pos_St.loc[rownum+1] = 0     # next day Pos_St = 0  
    df.Trade_Tag.loc[rownum] = "u2j"
    P_L = ((Ex_price3 - df.close.loc[rownum-1]) + (Ex_price4 - df.close.loc[rownum-1])) / tick * tick_val       
    df.P_L.loc[rownum] = P_L
    rownum += 1       
    while df.TrendDay.loc[rownum] != 0:
        if df.FL.loc[rownum] > df.FL.loc[rownum-1]:
            if df.FL.loc[rownum-1] > df.FL.loc[rownum-2]:
                RE_CAN = True
                df.RE_CAN.loc[rownum] = RE_CAN
                df.Pos_St.loc[rownum] = 0
                df.RE_CAN.loc[rownum+1] = RE_CAN
                df.Pos_St.loc[rownum+1] = 0
                rownum += 1
                uptrend_zeros()      # call Pos_St = 0, RECAN=True scenario functions

            else:
                rownum += 1    
        else:
            rownum += 1        
       
           
""" Uptrend, Pos_St=2     c_type: open & d-o       case: high<c    low < a       """           
def u_2k():                                                                  # u2k:  No volatility impact - carry 2 positions 
    global rownum
    Pos_St = 2
    df.Pos_St.loc[rownum] = Pos_St
    df["Counter1"] = df["Counter1"].apply(pd.to_numeric)
    df["Counter2"] = df["Counter2"].apply(pd.to_numeric)
    df.Counter1.loc[rownum] = df.Counter1.loc[rownum-1] + 1
    df.Counter2.loc[rownum] = df.Counter2.loc[rownum-1] + 1 
    Pos_End = 2
    df.Pos_End.loc[rownum] = Pos_End
    df.Pos_St.loc[rownum+1] = 2
    df.Trade_Tag.loc[rownum] = "u2k"
    P_L = ((df.close.loc[rownum] - df.close.loc[rownum-1]) + (df.close.loc[rownum] - df.close.loc[rownum-1])) / tick * tick_val       
    df.P_L.loc[rownum] = P_L
    rownum += 1           
           
           
""" Uptrend, Pos_St=2     c_type: open & d-o       case: high < c      low > a       """           
def u_2m():                                                                 # u2m:  No volatility impact - carry 2 positions
    u_2k()       
    df.Trade_Tag.loc[rownum] = "u2m"       
   

""" Uptrend, Pos_St=2     c_type: DARK & d-d       case: high > c    low < a&b       """           
def u_2n():                                                              # u2n:  TP on 2 then RECAN on 2 then SL on 2
    global rownum
    Pos_St = 2
    df.Pos_St.loc[rownum] = Pos_St   
    """ TP on 2 and set RECAN=True  """
    Ex_price1 = df.c.loc[rownum] + tick
    Ex_price2 = df.c.loc[rownum] + tick
    Pos_End = 0
    RE_CAN = True
    """ Same day Re-entry due to Low exceeding 'd'  """
    En_price3 = df.d.loc[rownum] - tick   # RECAN
    En_price4 = df.d.loc[rownum] - tick   # RECAN
    Pos_End = 2                            # update Pos_End local variable 
    """ Same day SL on 2 due to Low exceeding 'b'   """
    Ex_price3 = df.b.loc[rownum] - tick   # Stop Loss on 1
    Ex_price4 = df.b.loc[rownum] - tick   # Stop Loss on 2
    Pos_End = 0                           # update Pos_End local variable   
    df.Pos_End.loc[rownum] = Pos_End      # final disposition of Pos_End local variable
    df.Pos_St.loc[rownum+1] = 0           # next day Pos_St = 0  
    RE_CAN = False
    df.RE_CAN.loc[rownum+1] = RE_CAN      # next day RECAN status is False
    df.Counter1.loc[rownum] = ""
    df.Trade_Tag.loc[rownum] = "u1n"     
    P_L = ((Ex_price1 - df.close.loc[rownum]) + (Ex_price3 - En_price3) + (Ex_price4 - En_price4)) / tick * tick_val
    df.P_L.loc[rownum] = P_L   
    df.Trade_Tag.loc[rownum] = "u2n"       
    rownum += 1 
    """ Check for Re-entry after Uptrend stop out """
    while df.TrendDay.loc[rownum] != 0:
        if df.FL.loc[rownum] > df.FL.loc[rownum-1]:
            if df.FL.loc[rownum-1] > df.FL.loc[rownum-2]:
                RE_CAN = True
                df.RE_CAN.loc[rownum] = RE_CAN
                df.Pos_St.loc[rownum] = 0
                df.RE_CAN.loc[rownum+1] = RE_CAN
                df.Pos_St.loc[rownum+1] = 0
                rownum += 1
                uptrend_zeros()      # call Pos_St = 0, RECAN=True scenario functions

            else:
                rownum += 1    
        else:
            rownum += 1
       
        
""" Uptrend, Pos_St=2     c_type: DARK & d-d       case: high > c    low < a         """           
def u_2p():                                                               # u2p:  TP on 2 then RECAN on 2 then carry 2
    global rownum
    Pos_St = 2
    df.Pos_St.loc[rownum] = Pos_St 
    """ TP on 2 and set RECAN=True  """
    Ex_price1 = df.c.loc[rownum] + tick
    Ex_price2 = df.c.loc[rownum] + tick
    Pos_End = 0
    RE_CAN = True
    """ Same day Re-entry due to Low exceeding 'd'  """
    En_price3 = df.d.loc[rownum] - tick   # RECAN
    En_price4 = df.d.loc[rownum] - tick   # RECAN
    Pos_End = 2                            # update Pos_End local variable 
    """ Carry 2 due to Low > 'b'   """  
    df.Pos_End.loc[rownum] = Pos_End      # final disposition of Pos_End local variable
    df.Pos_St.loc[rownum+1] = 2           # next day Pos_St = 2  
    RE_CAN = False
    df.RE_CAN.loc[rownum+1] = RE_CAN  # next day RECAN status is False
    df.Counter1.loc[rownum] = df.Counter1.loc[rownum-1] + 1
    df.Counter2.loc[rownum] = df.Counter2.loc[rownum-1] + 1
    df.Trade_Tag.loc[rownum] = "u1p"     
    P_L = ((Ex_price1 - df.close.loc[rownum]) + (df.close.loc[rownum] - En_price3) + (df.close.loc[rownum] - En_price4)) / tick * tick_val
    df.P_L.loc[rownum] = P_L     
    df.Trade_Tag.loc[rownum] = "u2p" 
    rownum += 1

           
""" Uptrend, Pos_St=2     c_type: DARK & d-d       case: high > c    low > a         """           
def u_2q():                                                               # u2q:  TP on 2 positions then set RECAN
    u_2g()             
    df.Trade_Tag.loc[rownum] = "u2q"       
           
""" Uptrend, Pos_St=2     c_type: DARK & d-d       case: high < c    low < a&b         """           
def u_2r():                                                                         # u2r:  SL on 2 positions
    u_2j()       
    df.Trade_Tag.loc[rownum] = "u2r"       
           
""" Uptrend, Pos_St=2     c_type: DARK & d-d       case: high < c    low < a           """           
def u_2s():                                                          # u2s:  No volatility impact - carry 2 positions
    u_2k()
    df.Trade_Tag.loc[rownum] = "u2s"       
           
""" Uptrend, Pos_St=2     c_type: DARK & d-d       case: high < c    low > a           """           
def u_2t():                                                          # u2t:  No volatility impact - carry 2 positions
    u_2k()           
    df.Trade_Tag.loc[rownum] = "u2t"              
           

""" Uptrend, Pos 0--> Pos 2--> Pos 0     c_type: open & d-o       case: high > c    low < d & b           """           
def u_0f():                                                            # u0f: RECAN on 2 positions then SL on 2 positions
    global rownum
    Pos_St = 0
    df.Pos_St.loc[rownum] = Pos_St
    En_price1 = df.d.loc[rownum] - tick
    En_price2 = df.d.loc[rownum] - tick
    df.En_price1.loc[rownum] = En_price1
    df.En_price2.loc[rownum] = En_price2
    Ex_price3 = df.b.loc[rownum] - tick
    Ex_price4 = df.b.loc[rownum] - tick 
    df.Ex_price3.loc[rownum] = Ex_price3
    df.Ex_price4.loc[rownum] = Ex_price4
    df.Counter1.loc[rownum] = ""  
    df.Counter2.loc[rownum] = ""
    RE_CAN = False
    df.RE_CAN.loc[rownum+1] = RE_CAN    # next day RECAN potential is False
    Pos_End = 0
    df.Pos_End.loc[rownum] = Pos_End
    df.Pos_St.loc[rownum+1] = 0     # next day Pos_St = 0  
    df.Trade_Tag.loc[rownum] = "u0f"
    P_L = ((Ex_price3 - En_price1) + (Ex_price4 - En_price2)) / tick * tick_val
    df.P_L.loc[rownum] = P_L
    rownum += 1 
    """ Check for Re-entry after Uptrend stop out """
    while df.TrendDay.loc[rownum] != 0:
        if df.FL.loc[rownum] > df.FL.loc[rownum-1]:
            if df.FL.loc[rownum-1] > df.FL.loc[rownum-2]:
                RE_CAN = True
                df.RE_CAN.loc[rownum] = RE_CAN
                df.Pos_St.loc[rownum] = 0
                df.RE_CAN.loc[rownum+1] = RE_CAN
                df.Pos_St.loc[rownum+1] = 0
                rownum += 1
                uptrend_zeros()      # call Pos_St = 0, RECAN=True scenario functions

            else:
                rownum += 1    
        else:
            rownum += 1
  
           
""" Uptrend, Pos_St=0     c_type: open & d-o       case: high > c    low < d            """           
def u_0g():                                                       # u0g: RECAN on 2 positions --->  TP on 2  ---> set RECAN
    global rownum 
    Pos_St = 0
    df.Pos_St.loc[rownum] = Pos_St                    
    En_price1 = df.d.loc[rownum] - tick
    En_price2 = df.d.loc[rownum] - tick
    df.En_price1.loc[rownum] = En_price1
    df.En_price2.loc[rownum] = En_price2
    Ex_price1 = df.c.loc[rownum] + tick
    Ex_price2 = df.c.loc[rownum] + tick
    df.Ex_price1.loc[rownum] = Ex_price1
    df.Ex_price2.loc[rownum] = Ex_price2    
    RE_CAN = True
    df.RE_CAN.loc[rownum+1] = RE_CAN    # logic: Pos_End disposition includes setting RECAN = True for the next day
    df.Counter1.loc[rownum] = ""  
    df.Counter2.loc[rownum] = ""
    Pos_End = 0
    df.Pos_End.loc[rownum] = Pos_End
    df.Pos_St.loc[rownum+1] = 0     # next day Pos_St = 0
    df.Trade_Tag.loc[rownum] = "u0g"
    P_L = ((Ex_price1- En_price1) + (Ex_price2 - En_price2)) / tick * tick_val
    df.P_L.loc[rownum] = P_L
    rownum += 1                    

           
""" Uptrend, Pos_St=0     c_type: open & d-o       case: high > c    low > d            """           
def u_0h():                                                                          # u0h: No RECAN, stay RECAN=True
    global rownum
    Pos_St = 0
    df.Pos_St.loc[rownum] = Pos_St         
    if df.RE_CAN.loc[rownum] == True:
        df.RE_CAN.loc[rownum+1] = df.RE_CAN.loc[rownum]   # next day RECAN potential is same as today
    df.Counter1.loc[rownum] = ""  
    df.Counter2.loc[rownum] = ""
    Pos_End = 0
    df.Pos_End.loc[rownum] = Pos_End
    df.Pos_St.loc[rownum+1] = 0     # next day Pos_St = 0 
    df.Trade_Tag.loc[rownum] = "u0h"
    rownum += 1         
           
           
""" Uptrend, Pos_St=0     c_type: open & d-o       case: high < c    low < d&b            """           
def u_0j():                                                         # u0j: RECAN on 2 positions then SL on 2 positions
    u_0f()       
    df.Trade_Tag.loc[rownum] = "u0j"       
   


""" Uptrend, Pos_St=0     c_type: open & d-o       case: high < c    low < d            """           
def u_0k():                                                          # u0k: RECAN on 2 positions then CARRY 2 positions
    global rownum  
    Pos_St = 0
    df.Pos_St.loc[rownum] = Pos_St
    En_price1 = df.d.loc[rownum] - tick
    En_price2 = df.d.loc[rownum] - tick
    df.En_price1.loc[rownum] = En_price1
    df.En_price2.loc[rownum] = En_price2    
    df.Counter1.loc[rownum] = 1
    df.Counter2.loc[rownum] = 1
    RE_CAN = False
    df.RE_CAN.loc[rownum] = RE_CAN    
    Pos_End = 2
    df.Pos_End.loc[rownum] = Pos_End
    df.Pos_St.loc[rownum+1] = 2
    df.Trade_Tag.loc[rownum] = "u0k"
    P_L = ((df.close.loc[rownum] - En_price1) + (df.close.loc[rownum] - En_price2)) / tick * tick_val
    df.P_L.loc[rownum] = P_L
    rownum += 1
    

""" Uptrend, Pos_St=0     c_type: open & d-o       case: high < c    low > d               """           
def u_0m():                                                                      # u0m: No RECAN, stay RECAN=True
    u_0h()
    df.Trade_Tag.loc[rownum] = "u0m"
           
""" Uptrend, Pos_St=0     c_type: DARK & d-D       case: high > c    low < d&b               """           
def u_0n():                                                          # u0n: RECAN on 2 positions then SL on 2 positions
    u_0f()
    df.Trade_Tag.loc[rownum] = "u0n"
           
""" Uptrend, Pos_St=0     c_type: DARK & d-D       case: high > c    low < d                """           
def u_0o():                                                          # u0o: RECAN on 2 positions then CARRY 2 positions
    u_0k()
    df.Trade_Tag.loc[rownum] = "u0o"
           
""" Uptrend, Pos_St=0     c_type: DARK & d-D       case: high > c    low > d                """           
def u_0q():                                                          # u0q: No RECAN, stay RECAN=True
    u_0h()
    df.Trade_Tag.loc[rownum] = "u0q"
           
""" Uptrend, Pos_St=0     c_type: DARK & d-D       case: high < c    low < d&b                """           
def u_0r():                                                           # u0r: RECAN on 2 positions then SL on 2 positions
    u_0f()
    df.Trade_Tag.loc[rownum] = "u0r"
           
""" Uptrend, Pos_St=0     c_type: DARK & d-D       case: high < c    low < d                  """           
def u_0s():                                                          # u0s: RECAN on 2 positions then CARRY 2 positions
    u_0k()
    df.Trade_Tag.loc[rownum] = "u0s"
           
""" Uptrend, Pos_St=0     c_type: DARK & d-D       case: high < c    low > d                  """           
def u_0t():                                                          # u0t: No RECAN, stay RECAN=True 
    u_0h()
    df.Trade_Tag.loc[rownum] = "u0t"
              
           
           
```
